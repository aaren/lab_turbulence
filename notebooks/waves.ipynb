```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.signal as sig

import gc_turbulence as g
import sparse_dmd as dmd
%matplotlib inline

import matplotlib as mpl
mpl.rc('figure', figsize=(15, 3))
```

```python
index = 'r14_01_14a'
r = g.ProcessedRun(cache_path=g.default_processed + index + '.hdf5')

u = r.U[:, 10, :]
v = r.V[:, 10, :]
w = r.W[:, 10, :]

vertical_mean_u = np.mean(u, axis=0)
vertical_mean_v = np.mean(v, axis=0)
vertical_mean_w = np.mean(w, axis=0)

t = r.T[0, 10, :]
z = r.Z[:, 10, 0]
```

```python
plt.plot(t, u[10], label='height = 10')
plt.plot(t, u[20], label='height = 20')
plt.plot(t, u[30], label='height = 30')
plt.plot(t, vertical_mean_u, label='vertical mean')
plt.legend()
plt.title('streamwise velocity through time')
```

```python
plt.plot(t, v[10], label='height = 10')
plt.plot(t, v[20], label='height = 20')
plt.plot(t, v[30], label='height = 30')
plt.plot(t, vertical_mean_v, label='vertical mean')
plt.legend()
plt.title('cross-stream velocity through time')
```

```python
plt.plot(t, w[10], label='height = 10')
plt.plot(t, w[20], label='height = 20')
plt.plot(t, w[30], label='height = 30')
plt.plot(t, vertical_mean_w, label='vertical mean')
plt.legend()
plt.title('vertical velocity through time')
plt.ylim(-0.01, 0.01)
```

There are waves superimposed on the velocity data. They are visible in the sreamwise and vertical components of velocity. There appear to be two dominant frequencies. 

We would like to remove these waves from the dataset to allow us to focus on the gravity current.

Subtracting a wave from the data requires us to know both the frequency, phase and amplitude of the wave. We could just apply a notch filter at the wave frequency but we would lose information from inside the current as well. If we know frequency, phase and amplitude then we can just subtract the wave from the data, assuming superposition.
